```python
# Grouping each offer event for individual users to find the overall status of the offer for each user
user_grouped_transcript = event_dummies.groupby(by = ['person','value']).sum()
```

```python
user_grouped_transcript
```

```python
# The observations regarding the repeated offers for same user are being changed into a single event
user_grouped_transcript['event_offer completed'] = user_grouped_transcript['event_offer completed'].apply(lambda x: 1 if (x>0) else 0)
user_grouped_transcript['event_offer received'] = user_grouped_transcript['event_offer received'].apply(lambda x: 1 if (x>0) else 0)
user_grouped_transcript['event_offer viewed'] = user_grouped_transcript['event_offer viewed'].apply(lambda x: 1 if (x>0) else 0)
```

```python
# changing 'offer' and 'person' from indexes to columns of the dataframe
user_grouped_transcript['offer'] = user_grouped_transcript.index.get_level_values(1)
user_grouped_transcript['person'] = user_grouped_transcript.index.get_level_values(0)
user_grouped_transcript.index = np.arange(user_grouped_transcript.shape[0])
```

```python
user_grouped_transcript.head()
```

```python
# Merging the user profiles (profile df) into the transcript df
profile_transcript = pd.merge(user_grouped_transcript,profile,how = 'inner',left_on = 'person',right_on = 'id')
```

```python
profile_transcript.head()
```

```python
#Dropping less relevent columns
profile_transcript.drop(labels = ['time','id','became_member_on'],axis = 1,inplace = True)
```

```python
# Merging the offer information (portfolio df) into transcript df
complete_transcript = pd.merge(profile_transcript,portfolio,right_on = 'id',left_on = 'offer',how = 'inner')
```

```python
complete_transcript.drop(labels = 'id',axis = 1,inplace = True)
```

```python
# Making dummy columns for categorical variables gender and channels
complete_transcript = pd.get_dummies(complete_transcript,columns = ['gender'])
complete_transcript['web'] = complete_transcript['channels'].apply(lambda x: 1 if ("web" in x) else 0)
complete_transcript['email'] = complete_transcript['channels'].apply(lambda x: 1 if ("email" in x) else 0)
complete_transcript['mobile'] = complete_transcript['channels'].apply(lambda x: 1 if ("mobile" in x) else 0)
complete_transcript['social'] = complete_transcript['channels'].apply(lambda x: 1 if ("social" in x) else 0)
complete_transcript.drop(labels = 'channels',axis =1,inplace = True)
```

```python
# getting dummy columns for categorical variable 'offer_type'
complete_transcript = pd.get_dummies(complete_transcript,columns = ['offer_type'])
```

## To measure the accuracy of the model I will be using the classification_report function of the scikit-learn library.
## The main measure metric I intend on using is the recall. The main aim I have in this project is to use offers as a means of customer retention and I want to keep False Negatives(predicting a user is not going to complete an offer even though that is False) to a minimum to avoid losing active customers due to wrong predictions.

```python
def check_accuracy(pred_values, real_values):
    '''
    This function prints the comparison of real target values with predicted target values for individual columns
    Input:
    pred_values,real_values
    '''
    pred_values = pred_values.transpose()
    for i in range(pred_values.shape[0]):
        print(classification_report(real_values.iloc[:, i], pred_values[i]))
```

### Splitting the data into features and target variables for training and testing a model

```python
X = complete_transcript[['age','income','reward','difficulty','duration','gender_F','gender_O','gender_M','web','email','mobile','social','offer_type_bogo','offer_type_discount','offer_type_informational']]
```

```python
Y = complete_transcript[['event_offer completed','event_offer viewed']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size = 0.2,random_state = 4)
```

```python
# Training first model
model = MultiOutputClassifier(estimator = RandomForestClassifier())
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# Checking the result of first model
check_accuracy(y_pred,y_test)
```

### The above results are not good enough and the model needs to be updated

```python
# Tuning the above model hyperparameters using GridSearchCV
parameters = {'estimator__n_estimators':(100,150,200),'estimator__criterion': ["gini", "entropy"]}
model = MultiOutputClassifier(estimator = RandomForestClassifier())
tuned_model = GridSearchCV(model,parameters)
tuned_model.fit(x_train,y_train)
```

```python
y_pred_tuned = tuned_model.predict(x_test)
```

```python
# checking the results of the model after tuning
check_accuracy(y_pred_tuned,y_test)
```

### The hyperparameter tuning didn't have much improvement in the model 

```python
# Training the second model
updated_model = MultiOutputClassifier(estimator = XGBClassifier())
updated_model.fit(x_train,y_train)
```

```python
y_pred_updated = updated_model.predict(x_test)
```

```python
# Checking the accuracy of the second model
check_accuracy(y_pred_updated,y_test)
```

### The XGBClassifier showed a significant improvement over the RandomClassifierModel

```python
# Hyperparameter tuning of the new model
parameters = {'estimator__n_estimators':(100,150,200),'estimator__learning_rate':[0.1,0.2,0.3],'estimator__subsample':[0.5,1]}
updated_model = MultiOutputClassifier(estimator = XGBClassifier())
updated_tuned_model = GridSearchCV(updated_model,parameters)
updated_tuned_model.fit(x_train,y_train)
```

```python
y_pred_updated_tuned = updated_tuned_model.predict(x_test)
```

```python
#Checking the result of the tuned model
check_accuracy(y_pred_updated_tuned,y_test)
```

### The tuning made the model very slightly more accurate
